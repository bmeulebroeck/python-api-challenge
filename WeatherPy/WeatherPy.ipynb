```python
import pandas as pd
import requests
import random
import matplotlib.pyplot as plt
import scipy.stats as st

from api_keys import ow_key
from citipy import citipy
```

## Generate Cities List

```python
# Need to generate a list of latitude/longitude, then use CitiPy to find out the nearest city to each pair
latitude = random.sample(list(range(-90,90)),25)
longitude = random.sample(list(range(-180,180)),25)
latitude, longitude
```

```python
# Combine into lat/long pairs
latlongpair = []
latlongpair = zip(latitude,longitude)
latlongpair
```

```python
city = citipy.nearest_city(-81, -29)
city.city_name
```

```python
# Create a list to hold the cities found
cities = []

# Use CitiPy to loop through the lats/longs to find nearby cities and add to list
for latlongpair in latlongpair:
    city = citipy.nearest_city(latlongpair[0], latlongpair[1]).city_name
    if city not in cities:
        cities.append(city)
        
len(cities)
```

```python
print(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Set up url to connect to OW API
url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = f"{url}appid={ow_key}&units=imperial&q="
```

```python
# Set up lists to collect the weather data for each city
temps = []
humidity = []
clouds = []
windspeed = []
lat = []
long = []
country = []
date = []

citiesfound = []

for city in cities:
    response = requests.get(query_url + city).json()
    
    try:
        lat.append(response['coord']['lat'])
        long.append(response['coord']['lon'])
        temps.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        clouds.append(response['clouds']['all'])
        windspeed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        citiesfound.append(city)
        print(f"{city} found, collecting info")
        
    except:
        print(f"{city} not found, moving to next city")
        pass
    
print(f"The latitude info rec: {lat}")
print(f"The longitude info rec: {long}")
print(f"The temperature info rec: {temps}")
print(f"The humidity info rec: {humidity}")
print(f"The cloudiness info rec: {clouds}")
print(f"The windspeed info rec: {windspeed}")
print(f"The country info rec: {country}")
print(f"The date info rec: {date}")
    
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_info = {"City":citiesfound, "Country":country, "Lat":lat, "Long":long, "Temp":temps, "Humidity":humidity, "Cloudiness":clouds, "Wind Speed":windspeed, "Date":date}
weather_info_df = pd.DataFrame(weather_info)
weather_info_df.to_csv('city_weather_info.csv', index=False)
weather_info_df
```

## Linear Regression

```python
# Split the info into two dataframes for North and South hemispheres
northern_df = weather_info_df.loc[weather_info_df["Lat"]>=0]
southern_df = weather_info_df.loc[weather_info_df["Lat"]<0]
```

```python
#northern_df
southern_df
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Linear regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(northern_df["Lat"], northern_df["Temp"])
regress_values = northern_df["Lat"]*slope+intercept
line_eq = "y = "+ str(round(slope,2)) + "x + " + str(round(intercept,2))

# Scatter plot with linear regression applied
plt.scatter(northern_df["Lat"], northern_df["Temp"])
plt.plot(northern_df["Lat"], regress_values, "r-")
plt.annotate(line_eq,(20,15), color='red', size=15)
plt.title("Temperature vs Latitude, Northern hemisphere")
plt.xlabel("Latitude")
plt.ylabel("Temperature")
#plt.grid(True)
plt.show()
```
